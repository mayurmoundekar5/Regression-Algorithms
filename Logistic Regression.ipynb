```python
import pandas as pd
import numpy as np
```

```python
claimants = pd.read_csv("F:\\PROGRAMMING\\R\\Exmaple\\Examples R\\Logistic Regression\\claimants.csv",sep=",")
```

```python
claimants
```

```python
import statsmodels.formula.api as sm
```

```python
logit_model = sm.logit('ATTORNEY~CLMAGE+LOSS+CLMINSUR+CLMSEX+SEATBELT',data = claimants).fit()
```

```python
logit_model.summary()
```

```python
logit_model.params
```

```python
from sklearn.metrics import confusion_matrix
import scipy
from sklearn import linear_model
```

```python
predict = logit_model.predict(pd.DataFrame(claimants[['CLMAGE','LOSS','CLMINSUR','CLMSEX','SEATBELT']]))
predict
```

```python
pcnf_matrix = confusion_matrix(claimants['ATTORNEY'],predict > 0.5)
pcnf_matrix
```

```python
from sklearn.metrics import accuracy_score
```

```python
Accuracy_Score = accuracy_score(claimants['ATTORNEY'],predict > 0.5)
Accuracy_Score
```
